```python
import pandas as pd
import os
import cv2
```

```python
#path dove si trova il dataframe hmdb51
path = "./data/hmdb51"
actions = os.listdir(f"{path}/videos")
```

```python
split1 = pd.read_csv(f"{path}/annotations/split1.csv")
```

```python
for parola in ["train", "test"]:
    newpath = f'{path}/rawframes/{parola}' 
    if not os.path.exists(newpath):
        os.makedirs(newpath)
```

```python
list_videos_action = []
for action in actions: 
    for parola in ["train", "test"]:
        newpath = f'{path}/rawframes/{parola}/{action}' 
        if not os.path.exists(newpath):
            os.makedirs(newpath)
    list_videos_action.extend(os.listdir(f"{path}/videos/{action}"))
    n_video = len(list_videos_action)
    counter_n_video = 0
```

```python
for action in actions: 
    for parola in ["train", "test"]:
        newpath = f'{path}/rawframes/{parola}/{action}' 
        if not os.path.exists(newpath):
            os.makedirs(newpath)
    list_videos_action = os.listdir(f"{path}/videos/{action}")
    n_video = len(list_videos_action)
    counter_n_video = 0
    for video in list_videos_action:
        if split1[split1["title"] == video]["train/test"].item() == 1:
            newpath = f'{path}/rawframes/train/{action}/{video}'
            if not os.path.exists(newpath):
                os.makedirs(newpath)
        elif split1[split1["title"] == video]["train/test"].item() == 2:
            newpath = f'{path}/rawframes/test/{action}/{video}'
            if not os.path.exists(newpath):
                os.makedirs(newpath)
        else:
            continue
        
        counter_n_video += 1
        video_reader = cv2.VideoCapture(f"{path}/videos/{action}/{video}")
        ret, frame = video_reader.read()
        count_frame = 1
        while ret:
            cv2.imwrite(os.path.join(f'{newpath}' , f'{video}frame{count_frame}.jpg'), frame)
            #cv2.waitKey(0)
            for i in range(4):          #leggo un frame ogni 4 
                ret, frame = video_reader.read()
            if not ret:
                break
            print(f'Action: {action}, video letti: {counter_n_video} su {n_video}')
            count_frame += 4
```
